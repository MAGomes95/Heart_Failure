```python
import os
from joblib import load

import catboost
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
```

#### __Data and Model Loading__

```python
validation = pd.read_csv(
    '../data/silver/tree_based/validation.csv',
    index_col=0
)

catboost_model = load(
    '../models/trained/cb.joblib'
)
```

#### __Predictions Generation and Primal Analysis__

```python
validation['predictions'] = catboost_model.predict(validation[np.setdiff1d(validation.columns, 'target')])
```

```python
validation['predictions'] = validation['predictions'] == validation['target']
```

```python
validation['predictions'].value_counts() / validation.shape[0]
```

```python
validation.loc[validation['predictions'] == 0, 'target'].value_counts() / validation.loc[validation['predictions'] == 0].shape[0]
```

#### __Feature Analysis__ 

Let's start by exploring the original features of the dataset.

```python
 discrete_original_features = [
    feature for feature in np.setdiff1d(validation.columns, ['predictions', 'target']) 
    if len(feature.split(' ')) == 1 and validation[feature].nunique() == 2
]

continuous_original_features = [
    feature for feature in np.setdiff1d(validation.columns, ['predictions', 'target']) 
    if len(feature.split(' ')) == 1 and validation[feature].nunique() > 2
]

interaction_features = np.setdiff1d(validation.columns, np.union1d(discrete_original_features, continuous_original_features))

interaction_features = np.setdiff1d(interaction_features, ['predictions', 'target'])
```

```python
for feature in continuous_original_features:
    
    plot = sns.displot(
        data=validation,
        x=feature,
        hue='predictions',
        kind='kde',
        palette='Set2'
    )
    
    plt.show()  
```

```python
for feature in discrete_original_features:
    
    to_plot = pd.DataFrame(
                data=np.transpose(
                    [[0, 0, 1, 1], [0, 1, 0, 1], [

                        validation.loc[(validation[feature] == 0) & (validation['predictions'] == 0)].shape[0]
                        / validation.loc[validation[feature] == 0].shape[0],

                        validation.loc[(validation[feature] == 0) & (validation['predictions'] == 1)].shape[0]
                        / validation.loc[validation[feature] == 0].shape[0],

                        validation.loc[(validation[feature] == 1) & (validation['predictions'] == 0)].shape[0]
                        / validation.loc[validation[feature] == 1].shape[0],

                        validation.loc[(validation[feature] == 1) & (validation['predictions'] == 1)].shape[0]
                        / validation.loc[validation[feature] == 1].shape[0]
                    ]]
                ),
                columns=[feature, 'classification', 'proportion']
            )

    plot = sns.barplot(
        x=feature,
        y='proportion',
        hue='classification',
        palette='Set2',
        data=to_plot
    )
    
    plt.show()
```

```python
for feature in interaction_features:
    
    plot = sns.displot(
        data=validation,
        x=feature,
        hue='predictions',
        kind='kde',
        palette='Set2'
    )
    
    plt.show()
```
