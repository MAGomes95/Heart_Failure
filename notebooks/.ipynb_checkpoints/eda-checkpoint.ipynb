**Data Loading and Arranging**

```python
data = pd.read_csv('../data/bronze/heart_failure_clinical_records_dataset.csv')

data.rename(columns={'DEATH_EVENT': 'target'}, inplace=True)
```

THe dataset has **299 instances aka patients charcaterized in 12 features**.

```python
discrete_vars = [feature for feature in data.columns if len(data[feature].unique()) == 2]

for discrete in discrete_vars: data[discrete] = data[discrete].astype('category')

continuous_vars = np.setdiff1d(data.columns, discrete_vars)
```

```python
for feature in data.columns:
    
    feature_data = data[feature]

    if feature in discrete_vars:
        
        plot = sns.catplot(
            x=feature, 
            data=data,
            kind='count',
            palette='Set2'
        )
        
    else:
        
        plot = sns.distplot(
            a=data[feature].values,
            kde=True
        )
        
    plot = plot.set(title=feature)
    
    plt.show()
```

**Statistical Analysis**

For further analysis, we will remove the time feature from our list of predictors.

The reason for this is that the variable is not related with the patient condition, given his defenition and can lead to erroneous conclusion

```python
continuous_stats = data[continuous_vars].describe()

continuous_stats.loc['skew'] = [data[feature].skew() for feature in continuous_vars]
continuous_stats.loc['interq dist'] = [data[feature].quantile(.75) - data[feature].quantile(.25) for feature in continuous_vars]
```

```python
outlier_prop = []

for feature in continuous_vars:
    
    first_quantile = continuous_stats.loc['25%', feature]
    third_quantile = continuous_stats.loc['75%', feature]
    iqr_distance = continuous_stats.loc['interq dist', feature]
    
    outlier_number = data[feature].loc[
        (data[feature] < first_quantile - 1.5*iqr_distance)
        & (data[feature] > third_quantile + 1.5*iqr_distance)
    ].shape[0]
    
    outlier_prop.append(outlier_number / 299)
    
continuous_stats.loc['outlier prop'] = outlier_prop
```

```python
continuous_stats
```

Interesting that at a first glance to the dataset, no feature exhibits outlier presence

Let's now perform a correlation and other analysis.

As we have for all features a large enought sample size, we can take the features distribution to be aproximately Gaussian and therefore we are taking the Pearson Correlation

```python
data[continuous_vars].corr(method='pearson')
```

In this first analysis, we can see that there's no evidence of strong correlations between the predictors.

Even thought there's seems not to be a significant correlation observed, let's take the covariates that display the most correlation and test their parcial correlation coefficient

```python
predictors = [
    ('serum_creatinine', 'age'), 
    ('serum_sodium', 'ejection_fraction'), 
    ('serum_creatinine', 'serum_sodium'),
]

for covars in predictors:
    x,y = covars
    
    others = list(np.setdiff1d(continuous_vars, [x, y]))

    analysis = pg.partial_corr(
        data=data,
        x=x,
        y=y,
        covar=others
    )
    
    print(f'Variables at test: {covars}')
    print(analysis)
    print('============')
```

We can see from the outputs of the partial correlation test, that all covariates that we tested exibit a weak but measurable linear relationship, having all the other predictors' efect eliminated.

Let's now focus on the discrete variables

```python
discrete_stats = data[discrete_vars].describe()

top_proportions = []

for feature in discrete_vars:
    
    top_frequency = discrete_stats.loc['freq', feature]
    
    total_count = discrete_stats.loc['count', feature]
    
    top_percentage = top_frequency / total_count
    
    top_proportions.append(top_percentage)
    
discrete_stats.loc['prop'] = top_proportions

discrete_stats
```

We can see that there is a class imbalacing problem on the target, and some too on the other features.

We will resume our analysis on the discrete features, by computing the proportion of observations fof each feature under each targets' class 

```python
for feature in discrete_vars:
    
    if feature is not 'target':
    
        plot = sns.catplot(x=feature, hue='target', data=data, kind='count', palette='Set2')
    
        plt.show()
```

From this plot we cannot take any particularly usefull information. Sure for most discrete features, we have a bigger count of a particular target class on one category from the compared with the other features' category, but this can simply because we have a greater predominance of that category in that feature.

In this sense, let's perform a chi squared test for independence between the discrete variables and the target and also through Naive Bayes theorem we will compute the conditional probabilities given the target and each of the discrete features.

```python
for feature in discrete_vars:
    
    if feature is not 'target':
    
        _, _, chi2_stats = pg.chi2_independence(data=data, x=feature, y='target')
    
        print(f'===== {feature} / target =====')
    
        print(chi2_stats)
        
        print('\n')
```

Indeed, the diabetes and sex features show a huge pvalue, and we might jump to the conclusion that they are independent form the target. But thinking again, the power of all the tests is very low and might be not rejecting the null hyphotesis when we should. Therefore, we can't conclude nothing from the above tests. 

```python
for feature in discrete_vars:
    
    if feature is not 'target':
        
        this_bayes = pd.DataFrame(
            columns=[f'{feature}_class', 'target_class', f'p({feature} and target)', 'p(target)', f'p(target|{feature})']
        )
    
        this_bayes[f'{feature}_class'] = [0, 0, 1, 1]
        this_bayes['target_class'] = [0, 1, 0, 1]
        
        for line in range(this_bayes.shape[0]):
                
            feature_class = this_bayes.loc[line, f'{feature}_class']
            target_class = this_bayes.loc[line, 'target_class']
        
            this_bayes.loc[line, f'p({feature} and target)'] = data.loc[
                (data['target'] == target_class) & (data[feature] == feature_class)
            ].shape[0] / data.shape[0]
        
            this_bayes.loc[line, 'p(target)'] = data.loc[
                data['target'] == target_class
            ].shape[0] / data.shape[0]
        
            this_bayes.loc[line, f'p(target|{feature})'] = this_bayes.loc[
                line, f'p({feature} and target)'
            ] / this_bayes.loc[line, 'p(target)']
        
        print(this_bayes)
        print('\n')
```

From this analysis is quite clear the probabilities associated with each of the targets' class does not vary significantly across each of the features' classes. In this sense doesn't seem to exisit a discrete feature that is a good predictor of the target, but note that the feature that seems to have a more discriminative power over the target is the high blood pressure and that the diabetes features seems to be quite useless.

Let's move with our analysis with the continuous variables.

In each of the continuous variables, we will perform a T test, given the targets' classes.

```python
for feature in continuous_vars:
    
    feature_target_0 = data.loc[data['target'] == 0, feature] 
    
    feature_target_1 = data.loc[data['target'] == 1, feature]
    
    _, pvalue = stats.ttest_ind(a=feature_target_0, b=feature_target_1, equal_var=False)
    
    print(f'{feature}: p_value is {pvalue}\n')
```

Interesting indeed! It seems that **age, ejection_fraction, serum_creatinine and serum_sodium are good predictors for the target**, in the sense, that the distribution os their values under each of the target class are significatilly different, considering a $\alpha$ of 0.05

Let's take a closer good to this situation

**Feature Enginnering**

Given the insight that I were able to acquire from the data, let's expand the dataset with the three mentioned above features.

In this sense, we will perform the quocient and product of the pairs of feature mentioned, and check through a hypothesis test, which best discrimates the targets' class.

```python
pairs = [
    ('age', 'ejection_fraction'),
    ('serum_creatinine', 'serum_sodium'),
    ('serum_creatinine', 'ejection_fraction')
]

for this_pair in pairs:
    
    element_0, element_1 = this_pair
    
    product_arr = data[element_0] * data[element_1]
    
    quocient_arr = data[element_0] / data[element_1]
    
    this_set = pd.DataFrame(
        data=np.transpose([
            data['target'] , product_arr.values, quocient_arr.values
        ]),
        columns=['target', 'product', 'quocient']
    )
    
    product_arr_0 = this_set.loc[this_set['target'] == 0, 'product']
    
    product_arr_1 = this_set.loc[this_set['target'] == 1, 'product']
    
    _, product_pvalue = stats.ks_2samp(product_arr_0, product_arr_1)
    
    quocient_arr_0 = this_set.loc[this_set['target'] == 0, 'quocient']
    
    quocient_arr_1 = this_set.loc[this_set['target'] == 1, 'quocient']
    
    _, quocient_pvalue = stats.ks_2samp(quocient_arr_0, quocient_arr_1)
    
    if product_pvalue < quocient_pvalue:
        
        data[f'{element_0}_x_{element_1}'] = product_arr
        
        plot = sns.displot(
            x='product',
            kind='kde',
            hue='target',
            palette='Set2',
            data=this_set
        )
        
    else:
        
        data[f'{element_0}_/_{element_1}'] = quocient_arr
        
        plot = sns.displot(
            x='quocient',
            kind='kde',
            hue='target',
            palette='Set2',
            data=this_set
        )
        
    plt.show()
```
